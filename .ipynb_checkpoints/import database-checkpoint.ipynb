```python
import pandas as pd
```

```python
df = pd.read_csv("test.csv")
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df
```

```python
obj = df.groupby('class')
```

```python
container=[]
backup = {}
for index,i in enumerate(obj.groups):
    temp = {
        "model": "home.classdetails",
        "pk": index+1,
        "fields": {
            "className": int(i),
            "classTitle": "1st class",
        }
    }
    if temp not in container:
        container.append(temp)
    obj1 = df.groupby('subject')
    for index1 , i1 in enumerate(obj1.groups):
        temp1 ={"model": "home.subject",
            "pk": index1+1,
            "fields": {
                "className": index+1,
                "SubjectName": i1,
                "SubjectTitle": i1,
            }
        }
        if temp1 not in container:
            container.append(temp1)
        obj2 = df.groupby('topic')
        for index2, i2 in enumerate(obj2.groups):
            temp2 = {
                "model": "home.topic",
                "pk": index2+1,
                "fields": {
                    "Subject": index1+1,
                    "TopicName": i2,
                    "topicTitle": i2,
                }
            }
            if temp2 not in container:
                container.append(temp2)
            obj3 = df.groupby('subtopic')
            for index3 , i3 in enumerate(obj3.groups):
                temp3 ={"model": "home.subtopic",
                        "pk":index3+1,
                        "fields": {
                            "topic": index2+1,
                            "SubTopicName": i3,
                            "subtopicTitle": "Present tense in first topic.",
                        }
                    }
                if temp3 not in container:
                    backup[i3] = index3+1
                    container.append(temp3)
```

```python
counter =1
for index, i in df.iterrows():
    temp = {"model": "home.explain", 
            "pk": counter, 
            "fields": {"SubTopic": backup[i['subtopic']],
                       "explaintitle": i['explains'],
                       "body":  ""
                       "imgBlack": i['bimage'],
                       "imgColor": i['image'],}}
    counter+=1
    container.append(temp)
```

```python
import json
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(container, f, ensure_ascii=False, indent=4)
```
