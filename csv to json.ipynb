```python
import pandas as pd
import json
```

```python
df = pd.read_csv("test1.csv")
```

```python
df.fillna(method='ffill', inplace=True)
```

```python

df
```

```python
def get_or_create(_dict, key):
    if key not in _dict:
        _dict[key] = dict()
    return _dict[key]


result = {}
for index, item in df.iterrows():
    cls_dict = get_or_create(result, item["class"])
    result[item["class"]] = cls_dict
    sub_dict = get_or_create(cls_dict, item["subject"])
    topic_dict = get_or_create(sub_dict, item["topic"])
    sub_topic_dict = get_or_create(topic_dict, item["subtopic"])
    explain_dict = get_or_create(sub_topic_dict, item["explains"])
    explain_dict["image"] = item["image"]
    explain_dict["bimage"] = item["bimage"]

global_list = []

cls_counter = 1
sub_couner = 1
topic_counter = 1
sub_topic_counter = 1
expl_counter = 1

for name_float, subjects in result.items():
    cls_item = {
        "model": "home.classdetails",
        "pk": cls_counter,
        "fields": {
            "className": f'name for {name_float}',
            "classTitle": f'title for {name_float}',
        }
    }
    global_list.append(cls_item)
    cls_counter += 1

    for sub_name, topics in subjects.items():
        sub_item = {
            "model": "home.subject",
            "pk": sub_couner,
            "fields": {
                "className": cls_counter - 1,
                "SubjectName": sub_name,
                "SubjectTitle": f'title for {sub_name}'
            }
        }
        global_list.append(sub_item)
        sub_couner += 1
        for topic_name, sub_topics in topics.items():
            topic_item = {
                "model": "home.topic",
                "pk": topic_counter,
                "fields": {
                    "Subject": sub_couner - 1,
                    "TopicName": topic_name,
                    "topicTitle": f'title for {topic_name}'
                }
            }
            global_list.append(topic_item)
            topic_counter += 1
            for sub_topic_name, explains in sub_topics.items():
                sub_topic_item = {
                    "model": "home.subtopic",
                    "pk": sub_topic_counter,
                    "fields": {
                        "topic": topic_counter - 1,
                        "SubTopicName": sub_topic_name,
                        "subtopicTitle": f'title for {sub_topic_name}'
                    }
                }
                global_list.append(sub_topic_item)
                sub_topic_counter += 1
                for expl_title, explain_items in explains.items():
                    expl_ = {
                        "model": "home.explain",
                        "pk": expl_counter,
                        "fields": {
                            "SubTopic": sub_topic_counter - 1,
                            "explaintitle": expl_title,
                            "imgBlack": explain_items["bimage"],
                            "imgColor": explain_items["image"],
                            "body": f' {expl_title}'
                        }
                    }
                    global_list.append(expl_)
                    expl_counter += 1

with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(global_list, f, ensure_ascii=False, indent=4)
    print("write completed!!!")
```
